#### HTML構造

HTMLの構造はタグと呼ばれる記号で構成されている。
タグには一般的に開始タグと終了タグがあり、記述をタグで囲むことにより文章構造を定義しています。この、タグで囲われた塊のことを「要素」といいます。

## 2.スクレイピングの概要

### クローリングとスクレイピング

#### Webを巡回する「クローリング」
クローリングは自動的にWebサイトを巡回して情報を収集することです。Webサイトを巡回するプログラムのことを「クローラー」や「スパイダー」とよびます。
#### 情報を削り取る「スクレイピング」
「スクレイピング」はWebサイトから必要な情報を抽出する処理のことを指します。（スクレイピングはもともと「こすりとる」「かき集める」という意味）
クローリングによって収集したデータの中から、必要な部分だけを抽出したり、加工することを指します。

多くの場合は、スクレイピングの作業は、ダウンロードしたHTMLを解析し、必要な情報を抽出することを指します。

### スクレイピングは悪いことか？
スクレイピングは字面のせいか、悪い印象を持たれがちですが、悪いものではありません。
ただし、利用規約・著作権への配慮・相手サーバーへの過剰な負荷を考慮する必要があります。

### スクレイピングで気をつけること

- 著作権をまもる
- アクセスし過ぎて業務妨害をしない
- クローリング禁止のところからクローリングしない

具体的にやること
- robots.txtファイルを確認

以下にやるべきステップをまとめています。
| ステップ | 確認事項 | 概要 |
| --- | --- | --- |
|1|ウェブサイトのURLを確認|スクレイピングを行うウェブサイトのURLを取得します。これがスクレイピングのスタートポイントとなります。|
|2|robots.txtファイルの存在を確認|ウェブサイトのURLに"/robots.txt"を追加してアクセスします。例えば、`https://www.example.com/`というURLのウェブサイトであれば、`https://www.example.com/robots.txt`にアクセスします。これにより、ウェブサイトの管理者が公開しているrobots.txtファイルを見つけることができます。|
|3|robots.txtファイルの内容を読む|robots.txtファイルを開き、その内容を確認します。このファイルはウェブクローラーがウェブサイトのどの部分にアクセスできるか、またはアクセスすべきでないかを示す規則を含んでいます。|
|4|`User-agent`と`Disallow`の行を探す|`User-agent`は特定のウェブクローラーを指定します。`User-agent: *`はすべてのウェブクローラーに対する規則を示します。`Disallow`行は特定のウェブクローラーがアクセスを避けるべきウェブサイトのパスを示します。`Disallow: /`はウェブサイトの全ページに対するアクセスが禁止されていることを示します。|
|5|ウェブサイトの利用規約を確認|robots.txtファイル以外にも、ウェブサイトの利用規約やプライバシーポリシーに目を通すことが重要です。これらの文書には、ウェブスクレイピングに関する追加の規則や制約が記載されている場合があります。|
|6|必要ならばウェブサイトの所有者に連絡を取る|robots.txtや利用規約が曖昧である場合や、あなたのスクレイピング目的がこれらの規則で禁止されている可能性がある場合、ウェブサイトの所有者や管理者に直接連絡を取って許可を得ることを検討してください。これにより、法的トラブルを避けることができます。|

※補足
1. **User-agentとは何か？**
User-agentはウェブクローラーやスクレイパーのことを指し、特にrobots.txtファイル内で指定されます。User-agentの行は、その後に続く指示（DisallowやAllowの行）が適用されるウェブクローラーを定義します。たとえば、`User-agent: Googlebot`はGoogleのウェブクローラー（Googlebot）に対する指示が次に続くことを意味します。特殊なケースとして`User-agent: *`があります。アスタリスク(*)はワイルドカードとして機能し、任意のウェブクローラーすべてに適用するという意味になります。

**1. User-agentでよく使われるもの:**
User-agentとしてよく使われるものは、そのウェブサイトに訪れる一般的なウェブクローラーやウェブブラウザーの種類によります。以下にいくつかの一般的なUser-agentの例を示します：

- `Googlebot`: Googleのウェブクローラー
- `Bingbot`: Microsoftの検索エンジンBingのウェブクローラー
- `Slurp`: Yahooのウェブクローラー
- `DuckDuckBot`: DuckDuckGoのウェブクローラー
- `Baiduspider`: Baiduのウェブクローラー（主に中国で使用）
- `YandexBot`: Yandexのウェブクローラー（主にロシアで使用）
- `*`: すべてのウェブクローラーを指すワイルドカード

注意すべき点は、具体的なクローラー名を指定した場合、その指示はそのクローラー専用であるということです。つまり、`User-agent: Googlebot`の下にある指示はGooglebotに対してのみ適用され、他のクローラーには影響しません。すべてのクローラーに対する指示を出したい場合は、`User-agent: *`を使用します。

2. **Disallowとは何か？**
Disallowは、特定のウェブクローラーがアクセスすべきでないウェブサイトのパス（URLの一部）を指定します。`Disallow: /private`という行がある場合、`/private`パスにあるすべてのページへのアクセスは禁止されているという意味になります。また、`Disallow: /`という指示がある場合、ウェブサイトのすべてのページへのアクセスが禁止されていると解釈します。一方、`Disallow:`とパス部分が空の場合、すべてのページへのアクセスが許可されていると解釈します。

**2. Disallowに含まれるパスはすべてクローリング禁止という意味であっているか？**
基本的にはそうです。`Disallow: /path/`と指定されている場合、そのパスとその下位のパスに対するクローリングやスクレイピングが禁止されていると解釈すべきです。しかし、robots.txtはあくまで自己申告的なものであり、悪意を持って無視するクローラーやスクレイパーからウェブサイトを保護する能力は限定的です。さらに、robots.txtの解釈についてはクローラーによって若干異なる可能性もあります。

また、`Disallow`行が存在しない場合、すべてのパスへのアクセスが許可されていると解釈されます。さらに、Disallow行が存在するものの、パス部分が空（`Disallow:`）の場合も、すべてのパスへのアクセスが許可されていると解釈されます。

3. **どのように解釈すればよいか？**
robots.txtファイルを読む際の基本的なルールは、上から下へと読むことです。まず、User-agent行を見つけ、そのクローラーが自分の使っているもの（または`*`が指定されている場合）と一致するかを確認します。その後、そのUser-agent行に続くDisallow行を見つけ、その行が示すパスに対するアクセスが禁止されているかどうかをチェックします。このようにして、自分がアクセスしようとしているURLがDisallowで指定されているかどうかを確認します。禁止されているパスにはアクセスしないよう注意しましょう。

## 3.スクレイピングにチャレンジ

### まずはやってみる

最も簡単な例題をやってみます。

https://www.ymori.com/books/python2nen/test1.html

こちらのページを参考にスクレイピングをしていきます。
<img src="img/img036.png">

```python
#まずは今回使うrequestsとBeautifulSoupをインポートします。
import requests
from bs4 import BeautifulSoup

# Webページを取得して解析する
load_url = "https://www.ymori.com/books/python2nen/test1.html"

html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")

# title、h2、liタグを検索して表示する
# find()は最初の1つを取得する
# find_all()は全てを取得する
print("title部分:", soup.find("title"))
print("H2部分:",soup.find("h2"))
print("li部分(1要素):",soup.find("li"))
print("li部分(全要素):",soup.find_all("li"))

# selectで同じことをやってみる
print("title部分:", soup.select_one("title"))
print("H2部分:",soup.select_one("h2"))
print("li部分(1要素):",soup.select_one("li"))
print("li部分(全要素):",soup.select("li"))
```

### スクレイピングのチートシート

### beautifulsoupでよく使うメソッド
beautifulsoupでよく使うメソッドは以下です。
| メソッド | 引数 | 説明 |
| --- | --- | --- |
| find() | 検索するHTMLタグ | 引数に一致する最初の1つの要素を取得します。 |
| find_all() | 検索するHTMLタグ | 引数に一致する全ての要素を取得します。 |

| select_one() | CSSセレクタ | CSSセレクタに一致する最初の1つの要素を取得します。 |

そしてよく使うのはselectです。

### findとselectの違い

find()とfind_all()は、指定したHTMLタグに一致する要素を検索します。一方、select()とselect_one()は、指定したCSSセレクタに一致する要素を検索します。それぞれのメソッドは、最初の1つの要素を取得するか、一致する全ての要素を取得するかによって異なります

| find_all() | select() | やっていること |
| --- | --- | --- |
| soup.find_all("li") | soup.select("li") | タグがliのすべての要素 |
| soup.find_all(class_="ramen") | soup.select(".ramen") | クラスがramenのすべての要素 |
| soup.find_all(class_="ramen item") | soup.select("[class='ramen item']") | クラスが完全に'ramen item'一致するすべての要素 |
| ー | soup.select(".ramen.item") | クラスが'ramen'と'item'を含むすべての要素 |
| soup.find_all("li", class_="favorite") | soup.select("li.favorite") | タグがliで、クラスにfavoriteを含むすべての要素 |
| soup.find_all(id="miso") | soup.select("#miso") | idがmisoのすべての要素 |
| soup.find_all(id=["miso", "tonkotsu"]) | soup.select("#miso, #tonkotsu") | idがmisoまたはtonkotsuのすべての要素 |
| soup.find_all('a', {'class': 'ramen'}) | soup.select('a.ramen') | タグがaで、クラスがramenのすべての要素 |
| soup.find_all(['a', 'p']) | soup.select('a, p') | タグがaまたはpのすべての要素 |

```python
import requests, os, time, json
import urllib.parse
from bs4 import BeautifulSoup

# 初期設定 --- (※1)
shodou_url = 'https://uta.pw/shodou/index.php?master'
save_dir = os.path.join(os.path.dirname(__file__), 'images')
logfile = 'images.json'

# 書道掲示板の画像をダウンロード --- (※2)
def download_shodou(target_url):
    # 保存先のディレクトリがなければ作成 --- (※3)
    if not os.path.exists(save_dir):
        os.mkdir(save_dir)
    # HTMLをダウンロード --- (※4)
    html = requests.get(target_url).text
    time.sleep(1) # アクセスしたら待機 --- (※5)
    # HTMLを解析 --- (※6)
    soup = BeautifulSoup(html, 'html.parser')
    # 画像一覧が配置されている要素一覧を取得 --- (※7)
    a_div = soup.select('#contents_body > div')
    if len(a_div) == 0:
        print('[エラー] 要素の取得に失敗')
        return
    images = []
    # 抽出範囲からさらに画像の一覧を抽出 --- (※8)
    for img in a_div[0].find_all('img'):
        # <img src="xxx">のsrcを取得 --- (※9)
        src = img.attrs['src']
        alt = img.attrs['alt']
        # 絶対URLに変換 --- (※10)
        a_url = urllib.parse.urljoin(target_url, src)
        # ファイル名を決めてダウンロード
        fname = os.path.join(save_dir, src.replace('/', '_'))
        download_to_file(a_url, fname)
        # データとして保存
        images.append({'title': alt, 'url': a_url, 'file': fname})
    # 作業内容をログとしてJSONに保存 --- (※11)
    with open(logfile, 'w', encoding='utf-8') as fp:
        json.dump(images, fp)

# 実際に画像をダウンロード --- (※12)
def download_to_file(url, file):
    print('download:', url)
    # コンテンツをダウンロード --- (※13)
    bin = requests.get(url).content
    time.sleep(1) # アクセスしたら待機
    # 画像ファイルへ保存 --- (※14)
    with open(file, 'wb') as fp:
        fp.write(bin)

if __name__ == '__main__':
    download_shodou(shodou_url)
```

#### チャレンジ３：バグトラッキング掲示板のテーブルデータを取得
チャレンジの目的
- 複数ページに分かれた掲示板のデータを巡回して全部抽出してみる
- 集計グラフを作ってみる

やること
- 掲示板にかかれているデータがどのような構造か？
- 掲示板ログの続きのページのリンクはどうなっているのか？

<img src="img/img039.png">

```python
import requests, os, time, json
import urllib.parse
from bs4 import BeautifulSoup

# 初期設定 --- (※1)
BBS_URL = 'https://nadesi.com/cgi/bug/index.php'
LOGFILE = 'bbs_logs.json'
MAX_PAGES = 5 # 最大ページ数
logs = [] # 収集済みログデータ保存用
pages = [] # ダウンロード済みページ管理用

# 掲示板にアクセスしてデータを取り出す --- (※2)
def get_logs(target_url):
    # 最大ページ数の確認 --- (※3)
    if len(pages) > MAX_PAGES:
        return # 最大ページ数を超えたなら戻る
    # 二重にページを取得していないかチェック --- (※4)
    if target_url in pages:
        return # 既にダウンロード済みなら戻る
    pages.append(target_url)
    # HTMLをダウンロード --- (※5)
    html = requests.get(target_url).text
    time.sleep(1) # アクセスしたら待機
    # HTMLを解析 --- (※6)
    soup = BeautifulSoup(html, 'html.parser')
    # 掲示板のログデータを抽出 --- (※7)
    for row in soup.select('#body div.thread > table tr'):
        # trの下のtd要素を抽出 --- (※8)
        td_list = list(row.children)
        # ログページのURLを取得 --- (※9)
        a = td_list[0].find('a')
        if a is None: continue
        # ログのURLを絶対URLに変換 --- (※10) 
        href = a.attrs['href']
        href = urllib.parse.urljoin(target_url, href)
        # ログの各種情報を辞書型に入れる --- (※11)
        info = {
            'id': td_list[0].text,
            'title': td_list[1].text,
            'date': td_list[3].text,
            'priority': td_list[4].text,
            'status': td_list[5].text,
            'link': href,
        }
        print(info['id'], info['title'], info['link'])
        logs.append(info) # ログに追加
    # 次へボタンのリンクを求める --- (※12)
    for e in soup.select('.pager > a'):
        if e.text != '次へ→': continue
        link = e.attrs['href']
        # リンクを絶対URLに変換
        link = urllib.parse.urljoin(target_url, link)
        # 再帰的に掲示板の内容をダウンロード --- (※13)
        get_logs(link)

def save_logs():
    # ログの内容をファイルに保存 --- (※14)
    with open(LOGFILE, 'w', encoding='UTF-8') as fp:
        json.dump(logs, fp, ensure_ascii=False)
    print('ログの数:', len(logs))

if __name__ == '__main__':
    get_logs(BBS_URL) # データを取得
    save_logs() # データを保存
```

ではとってきたデータからグラフを作成してみます

```python
import json, japanize_matplotlib
import matplotlib.pyplot as plt

# JSONを読む --- (※1)
data = json.load(open('bbs_logs.json', encoding='utf-8'))
# データを数える --- (※2)
priority = {}
status = {}
for row in data:
    pr = row['priority']
    st = row['status']
    if pr not in priority: priority[pr] = 0
    if st not in status: status[st] = 0
    priority[pr] += 1
    status[st] += 1
print(priority)
print(status)
# グラフを描画 --- (※3)
fig = plt.figure()
# 優先度の円グラフを描画
ax1 = fig.add_subplot(1, 2, 1)
values = [v for _,v in priority.items()]
labels = [k for k,_ in priority.items()]
ax1.pie(values, labels=labels, autopct="%.1f")
ax1.set_title('優先度')
# 状態の円グラフを描画
ax2 = fig.add_subplot(1, 2, 2)
values = [v for _,v in status.items()]
labels = [k for k,_ in status.items()]
ax2.barh(labels, values)
ax2.set_title('状態')
plt.legend()
plt.show()
```

優先度ごとに解決・未解決を調べてグラフに絵画するプログラムは次のようになります。

```python
import json, japanize_matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# JSONを読む --- (※1)
data = json.load(open('bbs_logs.json', encoding='utf-8'))
# 以下のデータを埋めるように数える --- (※2)
result = {
    '緊急': {'解決': 0, '未解決': 0}, 
    '高':{'解決':0, '未解決': 0}, 
    '中':{'解決':0, '未解決': 0}, 
    '低':{'解決':0, '未解決': 0}, 
}
for row in data:
    st = row['status']
    pr = row['priority']
    if st == '解決': result[pr]['解決'] += 1
    if st == '未処理' or st == '修正中' or st == '再修正依頼':
        result[pr]['未解決'] += 1
print(result)
# グラフを描画 --- (※3)
df = pd.DataFrame({
    '解決': [v['解決'] for k,v in result.items()],
    '未解決': [v['未解決'] for k,v in result.items()]
}, index=[k for k,_ in result.items()])
df.plot(kind="barh",stacked=True,figsize=(10,8))
plt.show()
```
